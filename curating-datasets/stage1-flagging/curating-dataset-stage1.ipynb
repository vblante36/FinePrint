```python
def get_url_list(filepath):
    # Open and read the file line by line
    with open(filepath, 'r') as file:
        lines = file.readlines()

    # Process each line
    urls = []
    for line in lines:
        # Split the line into two parts: the IP/URL and the JSON data
        parts = line.split('{')
        if len(parts) == 2:
            json_data = "{" + parts[1]
            dict = json.loads(json_data)
            urls.append(dict['url'])

    return urls
```
